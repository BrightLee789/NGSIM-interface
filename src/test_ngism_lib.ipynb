```python
start_time = 1113433135300 + 1000 * 60 * 60 
end_time = start_time + 1000 * 60 * 60 * 0.2
plt.figure(figsize = (16,9), dpi = 300)
for veh_ID, tmp_veh in ng.veh_dict.iteritems():
    if tmp_veh.vr_list[0].unixtime > start_time and tmp_veh.vr_list[-1].unixtime < end_time:
        if np.random.rand() < 0.7:
#             print len(tmp_veh.vr_list)
            if len(tmp_veh.vr_list) < 4000:
#                 break
                plt.plot(list(map(lambda x: (x.unixtime - start_time)/1000, tmp_veh.vr_list)), 
                             list(map(lambda x: x.y, tmp_veh.vr_list)), color = 'black')
#                 break
plt.xlabel('Time (Unit:seconds)')
plt.ylabel('Space (Unit:feet)')
plt.ylim([100,1600])
plt.xlim([0, 700])
plt.show()
```
